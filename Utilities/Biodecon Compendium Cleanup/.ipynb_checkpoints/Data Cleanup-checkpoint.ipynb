```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('BioDeconProcessed_Original.xlsx', sheet_name = 'ProcessedData')

surfacesMaster = pd.read_excel('SurfacesMaster.xlsx', sheet_name = 'Sheet2')
```

```python
# pulling in the representative surfaces from the surface master and adding the column to the main data set
df['RepSurface'] = surfacesMaster.RepSurface

df.head()
```

```python
# clean numerical columns of strings and convert them to floats
strings = ['unlisted', 'undefined', 'not reported', 'Not reported', 'not applicable', 'Not Applicable', 'ambiguous', 'Ambiguous',
          'authors censor', 'inconclusive', 'NA', 'nd', 'No Value', 'not significant', 'NT', 'NR', '0.5 LR', 'as needed to keep wetted', '100-361',
          'Not significant', '>7', '>6', '50-75']

for i in strings:
    df.loc[df.LoadingNum == i, 'LoadingNum'] = np.nan
    df.loc[df.CoupArea == i, 'CoupArea'] = np.nan
    df.loc[df.Eff == i, 'Eff'] = np.nan
    df.loc[df.LPosRec == i, 'LPosRec'] = np.nan
    df.loc[df.TempNum == i, 'TempNum'] = np.nan
    df.loc[df.RHNum == i, 'RHNum'] = np.nan
    df.loc[df.Reapp == i, 'Reapp'] = np.nan
    
df['LoadingNum'] = df.LoadingNum.astype(float)
df['CoupArea'] = df.CoupArea.astype(float)
df['Eff'] = df.Eff.astype(float)
df['LPosRec'] = df.LPosRec.astype(float)
df['TempNum'] = df.TempNum.astype(float)
df['RHNum'] = df.RHNum.astype(float)
df['Reapp'] = df.Reapp.astype(float)
```

```python
# create a smaller data frame with only the columns we want so we can further clean and convert this subset of data
# this isn't necessary but dealing with the entire dataset stresses me out because of the column order
cd = pd.DataFrame()

cd['Reference'] = df.Ref
cd['AppMethod'] = df.AppMeth
cd['Surface'] = df.RepSurface
cd['Loading'] = df.LoadingNum
cd['LoadingUn'] = df.LoadingUn
cd['PosRec'] = 10 ** df.LPosRec 
cd['Area'] = df.CoupArea
cd['AreaUn'] = df.CoupAreaUn
cd['ConcDose'] = df.ConcDose
cd['ConcDoseUn'] = df.ConcDoseUn
cd['Temp'] = df.TempNum
cd['RH'] = df.RHNum
cd['ContTime'] = df.ContTimeNumMin
cd['Eff'] = df.Eff
cd['EffMeas'] = df.EffMeas
cd['Reapp'] = df.Reapp

cd.head()
```

```python
# clean up inconsistent unit naming just to make life a little easier because we all deserve that
cd.loc[(cd.AppMethod == 'liquid (ambiguous)'),'AppMethod'] = 'liquid ambiguous'
cd.loc[(cd.EffMeas == 'Qual Pos'), 'EffMeas'] = 'QualPos'
cd.loc[(cd.EffMeas == 'Qualneg'), 'EffMeas'] = 'QualNeg'
cd.loc[(cd.EffMeas == 'LRsurf'), 'EffMeas'] = 'LRSurf'
cd.loc[(cd.EffMeas == 'Lsurv'), 'EffMeas'] = 'LSurv'
cd.loc[(cd.AreaUn == 'inch^2'), 'AreaUn'] = 'in^2'
cd.loc[(cd.LoadingUn == 'spores/L'), 'LoadingUn'] = 'CFU/L'
```

```python
# converting all of the areas/volumes to either cm^2 or cm^3 so we have a standard
# some of the areas are given in mass so we're going to convert all mass to g 
cd.loc[(cd.AreaUn == 'mL'),'AreaUn'] = 'cm^3'

cd.loc[(cd.AreaUn == 'L'),'Area'] = cd['Area'] * 1000
cd.loc[(cd.AreaUn == 'L'),'AreaUn'] = 'cm^3'

cd.loc[(cd.AreaUn == 'gallon'),'Area'] = cd['Area'] * 3785.41
cd.loc[(cd.AreaUn == 'gallon'),'AreaUn'] = 'cm^3'

cd.loc[(cd.AreaUn == 'm^3'),'Area'] = cd['Area'] * 1000000
cd.loc[(cd.AreaUn == 'm^3'),'AreaUn'] = 'cm^3'

cd.loc[(cd.AreaUn == 'uL'),'Area'] = cd['Area'] * 0.001
cd.loc[(cd.AreaUn == 'uL'),'AreaUn'] = 'cm^3'

cd.loc[(cd.AreaUn == 'm^2'),'Area'] = cd['Area'] * 10000
cd.loc[(cd.AreaUn == 'm^2'),'AreaUn'] = 'cm^2'

cd.loc[(cd.AreaUn == 'mm^2'),'Area'] = cd['Area'] * 0.01
cd.loc[(cd.AreaUn == 'mm^2'),'AreaUn'] = 'cm^2'

cd.loc[(cd.AreaUn == 'ft^2'),'Area'] = cd['Area'] * 929.03
cd.loc[(cd.AreaUn == 'ft^2'),'AreaUn'] = 'cm^2'

cd.loc[(cd.AreaUn == 'in^2'),'Area'] = cd['Area'] * 6.4516
cd.loc[(cd.AreaUn == 'in^2'),'AreaUn'] = 'cm^2'

cd.loc[(cd.AreaUn == 'kg'),'Area'] = cd['Area'] * 1000
cd.loc[(cd.AreaUn == 'kg'),'AreaUn'] = 'g'
```

```python
# converting all of the loadings to CFU/cm^2 or CFU/cm^3
# we also need to convert the PosRec at the same time since it is the log of the same units and then take the log for LPosRec
cd.loc[(cd.LoadingUn == 'CFU/L'),'Loading'] = cd['Loading'] / 1000
cd.loc[(cd.LoadingUn == 'CFU/L'),'PosRec'] = cd['PosRec'] / 1000
cd.loc[(cd.LoadingUn == 'CFU/L'),'LoadingUn'] = 'CFU/cm^3'

cd.loc[(cd.LoadingUn == 'CFU/100cm^2'),'Loading'] = cd['Loading'] / 100
cd.loc[(cd.LoadingUn == 'CFU/L'),'PosRec'] = cd['PosRec'] / 100
cd.loc[(cd.LoadingUn == 'CFU/100cm^2'),'LoadingUn'] = 'CFU/cm^2'

cd.loc[(cd.LoadingUn == 'CFU/mL'),'LoadingUn'] = 'CFU/cm^3'
cd.loc[(cd.LoadingUn == 'CFU/ml'),'LoadingUn'] = 'CFU/cm^3'

# we also need to convert the log(CFU) into just CFU and then divide that by the coupon area for those that give an area
cd.loc[(cd.LoadingUn == 'log(CFU)'),'Loading'] = 10 ** cd['Loading']
cd.loc[(cd.LoadingUn == 'log(CFU)'),'LoadingUn'] = 'CFU'
```

```python
# dividing all of the Loadings and PosRecs by the area of the coupon for those with units of CFU so they can all be CFU/area
cd.loc[(cd.LoadingUn == 'CFU') & (pd.isnull(cd.Area) == False),'Loading'] = cd['Loading'] / cd['Area']
cd.loc[(cd.LoadingUn == 'CFU') & (pd.isnull(cd.Area) == False),'PosRec'] = cd['PosRec'] / cd['Area']
cd.loc[(cd.LoadingUn == 'CFU') & (cd.AreaUn == 'cm^3') & (pd.isnull(cd.Area) == False),'LoadingUn'] = 'CFU/cm^3'
cd.loc[(cd.LoadingUn == 'CFU') & (cd.AreaUn == 'cm^2') & (pd.isnull(cd.Area) == False),'LoadingUn'] = 'CFU/cm^2'
cd.loc[(cd.LoadingUn == 'CFU') & (cd.AreaUn == 'g') & (pd.isnull(cd.Area) == False),'LoadingUn'] = 'g'

# taking the log of the PosRec column and making it the new LPosRec
cd['LPosRec'] = np.log10(cd['PosRec'])
```
