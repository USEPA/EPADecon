```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in the entire dataset
master = pd.read_excel('..\Data Cleanup\cleaned_data.xlsx', sheet_name = 'master')
```

```python
concDoseParams = [
    ("foam spray", -1),
    ("liquid spray" , "UndergroundCarpet"),
    ("liquid immersion", "IndoorMisc"),
    ("fumigation", "HVAC"),
    ("liquid spray", "IndoorCarpet")
]
```

```python
def scatterPlotByDeconAgent(method, surface):
    df = master[(master['EffMeas'] == 'LR') & (master['ConcDoseUn'] == 'g/mL')]
    
    if method != -1 and surface != -1:
        df = df[(df['AppMethod'] == method) & (df[surface] == 1)]
        title = " ".join([word.capitalize() for word in method.split(" ")]) + ' and ' + surface
    elif method == -1 and surface != -1:
        df = df[(df[surface] == 1)]
        title = surface
    elif method != -1 and surface == -1:
        df = df[(df['AppMethod'] == method)]
        title = " ".join([word.capitalize() for word in method.split(" ")])
    
    indexNames = df[(np.isnan(df.Eff)) | (np.isnan(df.ConcDose))].index
    df.drop(indexNames, inplace=True)
    
    plt.figure(figsize=(11, 8))

    groups = df.groupby("DeconAgent")
    for name, group in groups:
        plt.plot(group["ConcDose"], group["Eff"], marker="o", linestyle="", label=name)
    
    plt.xlabel('ConcDose (g/mL)')
    plt.ylabel('Efficacy (LR)')
    plt.title(title)
    plt.legend()
    
    plt.show
    
    print(len(df))
    
for p in concDoseParams:
    scatterPlotByDeconAgent(*p)
```
